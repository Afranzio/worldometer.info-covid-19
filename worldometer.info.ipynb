```python
import pandas as pd

import requests

from bs4 import BeautifulSoup

# Set headers
headers = requests.utils.default_headers()

headers.update({ 'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'})

url = 'https://www.worldometers.info/coronavirus/'

response = requests.get(url,'lxml')

soup = BeautifulSoup(response.content, 'html.parser')

a = soup.find_all('td')
```

```python
index= []
country= []
total_cases= []
deaths = []
recovered= []
new_cases= []
new_deaths= []
active_cases = []
serious_cases = []
total_tests = []
```

```python

n = 120

while n<len(a):
    
    txt = a[n].text
    
    if txt == "":
        
        index.append(0)
        
    else:
        
        index.append(txt)
    
    n+=15
    
index = index[0:214]
```

```python
n1 = 121

while n1<len(a):
    
    txt = a[n1].text
    
    if txt == "":
        
        country.append(0)
        
    else:
        
        country.append(txt)
    
    n1+=15
    
country = country[0:214]
```

```python
n2 = 122

while n2<len(a):
    
    txt = a[n2].text
    
    if txt == "":
        
        total_cases.append(0)
        
    else:
        
        total_cases.append(txt)
    
    n2+=15
    
total_cases = total_cases[0:214]
```

```python
n3 = 123

while n3<len(a):
    
    txt = a[n3].text
    
    if txt == "":
        
        new_cases.append(0)
        
    else:
        
        new_cases.append(txt)
    
    n3+=15
    
new_cases = new_cases[0:214]
```

```python
n4 = 124

while n4<len(a):
    
    txt = a[n4].text
    
    if txt == "" or txt == " ":
        
        deaths.append(0)
        
    else:
        
        deaths.append(txt)
    
    n4+=15
    
deaths = deaths[0:214]
```

```python
n5 = 125

while n5<len(a):
    
    txt = a[n5].text
    
    if txt == "":
        
        new_deaths.append(0)
        
    else:
        
        new_deaths.append(txt)
    
    n5+=15
    
new_deaths = new_deaths[0:214]
```

```python

n6 = 126

while n6<len(a):
    
    txt = a[n6].text
    
    if txt == "":
        
        recovered.append(0)
        
    else:
        
        recovered.append(txt)
    
    n6+=15
    
recovered = recovered[0:214]
```

```python

n7 = 127

while n7<len(a):
    
    txt = a[n7].text
    
    if txt == "":
        
        active_cases.append(0)
        
    else:
        
        active_cases.append(txt)
    
    n7+=15
    
active_cases = active_cases[0:214]
```

```python

n8 = 128

while n8<len(a):
    
    txt = a[n8].text
    
    if txt == "":
        
        serious_cases.append(0)
        
    else:
        
        serious_cases.append(txt)
    
    n8+=15
    
serious_cases = serious_cases[0:214]
```

```python

n11 = 131

while n11<len(a):
    
    txt = a[n11].text
    
    if txt == "":
        
        total_tests.append(0)
        
    else:
        
        total_tests.append(txt)
    
    n11+=15
    
total_tests = total_tests[0:214]
```

```python
dic = {"country" : country, 'total_cases': total_cases,'new_cases': new_cases, 'deaths' : deaths,'new_deaths':new_deaths, 'recovered' : recovered, 'active_cases' : active_cases, 'serious_cases' : serious_cases, 'total_tests' : total_tests}

df = pd.DataFrame(dic)

df
```

```python
columns = ["total_cases",'new_cases',"deaths",'new_deaths',"recovered","serious_cases","active_cases", 'total_tests']

for column in columns:
    
    df[column] = df[column].str.replace(",","")

news = ['new_cases','new_deaths']

for new in news:
    
    df[new] = df[new].str.replace('+','')
    
df
```

```python
df.fillna(0,inplace = True)
df.replace('NaN', 0 ,inplace = True)
df.replace('N/A', 0 ,inplace = True)
df.replace(' ', 0 ,inplace = True)
df
```

```python
df['total_cases']  = df['total_cases'].astype(int)
df["deaths"] = df["deaths"].astype(int)
df['new_cases'] = df['new_cases'].astype(int)
df['new_deaths'] = df['new_deaths'].astype(int)
df['recovered'] = df['recovered'].astype(int)
df['active_cases'] = df['active_cases'].astype(int)
df['serious_cases'] = df['serious_cases'].astype(int)
df['total_tests'] = df['total_tests'].astype(int)
```
